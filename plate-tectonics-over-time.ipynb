### So how does this all work?
 
* ```pygplates``` is a software library that we use to reconstruct plates and plate boundaries back in time.
* ```matplotlib``` and its ```basemap``` plugin are libraries used to plot paleogeographic maps.
* The ```numpy``` library is used to do scientific computing.

A 'reconstruction' predicts where various features were located in the past. A 'feature' could be anything; a tectonic plate, a magnetic pick or a mineral deposit, for example. We can 'rewind' to see how the world looked in the past.

You can read the ```pygplates``` documentation at [http://www.gplates.org/docs/pygplates/](http://www.gplates.org/docs/pygplates/). Briefly:

* The ```pygplates.reconstruct()``` function carries out the reconstruction of features (coastlines, continental polygons, fracture zones, continent ocean boundaries, mineral deposits, magnetic picks.
* The ```pygplates.resolve_topologies()``` function plots the evolving network of plate boundaries (called plate topologies) at a given reconstruction time.
